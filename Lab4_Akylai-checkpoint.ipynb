# Оптическое распознавание набора данных рукописных цифр

```python
#Импорт
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

# Загрузка датасета
digits = load_digits()

# Изучение данных
print(digits.DESCR)
```

```python
# Значения ключа target_names
print("Unique Target Values:", set(digits.target))
# Размеры данных
print("Data Shape:", digits.data.shape)

# Вывод информации о целевых и признаковых данных
print("Target Names:", digits.target_names)
# Размеры данных
print("Target Shape:", digits.target.shape)
```

```python
# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)


# Вывод размеров полученных наборов данных
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(10, 4))
for i in range(10):
    axes[i // 5, i % 5].imshow(digits.images[i], cmap=plt.cm.gray_r, interpolation='nearest')
    axes[i // 5, i % 5].set_title(f"Label: {digits.target[i]}")
plt.show()
```

#### Этот код создает 5x2 сетку графиков, каждый из которых представляет собой гистограмму распределения значений пикселей для одной цифры (от 0 до 9). Гистограммы отображают, какие значения пикселей встречаются чаще всего.

```python
# Визуализация гистограмм распределения значений пикселей
fig, axes = plt.subplots(5, 2, figsize=(10, 12))
plt.suptitle("Distribution of Pixel Values in Digits Dataset", y=1.02)

for i in range(10):
    row, col = i // 2, i % 2
    axes[row, col].hist(digits.data[i], bins=10, color='skyblue', edgecolor='black')
    axes[row, col].set_title(f"Digit: {i}")

plt.tight_layout()
plt.show()
```

```python
# Построение модели k-ближайших средних
k = 3  # количество соседей
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)
```

```python
# Протестировать на тестовом наборе
y_pred = model.predict(X_test)

# Вывести прогноз модели и оценить качество
print("Прогноз модели:", y_pred)
```

```python
# Оценка качества модели
accuracy = metrics.accuracy_score(y_test, y_pred)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
# Вывод результатов
print("Набор данных с цифрами:")
print(f"Точность: {accuracy:.2%}")
print("Матрица ошибок:\n", confusion_matrix)
```

```python
import seaborn as sns

# Визуализация матрицы ошибок
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Выводы: Модель имеет высокую точность, но стоит проанализировать матрицу ошибок для выявления конкретных слабых сторон и возможных улучшений. Возможно, потребуется дополнительный анализ данных, балансировка классов или тюнинг параметров модели для улучшения её производительности в отношении определенных классов.

```python
# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.2, random_state=42)
```

```python
import pandas as pd

# Преобразование данных в DataFrame для удобства визуализации
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['target'] = wine.target


# Boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(x='target', y='alcohol', data=wine_df, palette='viridis')
plt.title('Boxplot of Alcohol Content for Wine Classes')
plt.xlabel('Wine Class')
plt.ylabel('Alcohol Content')
plt.show()

# Violin Plot
plt.figure(figsize=(12, 8))
sns.violinplot(x='target', y='alcohol', data=wine_df, palette='viridis')
plt.title('Violin Plot of Alcohol Content for Wine Classes')
plt.xlabel('Wine Class')
plt.ylabel('Alcohol Content')
plt.show()
```

В обоих случаях, графики позволяют вам сравнить распределение значений признака "алкоголь" между разными классами вин. Вы можете видеть, например, медианное значение, интерквартильный размах, а также различия в форме распределения. Это может быть полезным при анализе, чтобы лучше понять, какие признаки могут быть важны для различения классов вин.

```python
# Построение модели k-ближайших средних
k = 3  # количество соседей
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)
```

```python
# Протестировать на тестовом наборе
y_pred = model.predict(X_test)

# Вывести прогноз модели и оценить качество
print("Прогноз модели:", y_pred)
```

```python
# Оценка качества модели
accuracy = metrics.accuracy_score(y_test, y_pred)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
# Вывод результатов
print("Набор данных по распознаванию вин:")
print(f"Точность: {accuracy:.2%}")
print("Матрица ошибок:\n", confusion_matrix)
```

# Набор данных по раку груди висконсин (диагностический)

```python
#Импорты
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import seaborn as sns

# Загрузка датасета по раку груди
breast_cancer = load_breast_cancer()

# Изучение данных
print(breast_cancer.DESCR)
```

```python
# Уникальные значения целевой переменной
print("Unique Target Values:", set(breast_cancer.target))
# Размеры данных
print("Data Shape:", breast_cancer.data.shape)

# Вывод информации о целевых и признаковых данных
print("Feature Names:", breast_cancer.feature_names)
print("Target Names:", breast_cancer.target_names)
# Размеры данных
print("Target Shape:", breast_cancer.target.shape)
```

```python
# Разделение данных на обучающий и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.2, random_state=42)

```

```python
# Визуализация гистограмм распределения значений признаков
num_features = breast_cancer.data.shape[1]
num_rows = num_features // 2 + num_features % 2  # Calculate the number of rows needed

fig, axes = plt.subplots(num_rows, 2, figsize=(15, 3 * num_rows))
plt.suptitle("Distribution of Feature Values in Breast Cancer Dataset", y=1.02)

for i in range(num_features):
    row, col = i // 2, i % 2
    axes[row, col].hist(breast_cancer.data[:, i], bins=10, color='skyblue', edgecolor='black')
    axes[row, col].set_title(f"Feature: {breast_cancer.feature_names[i]}")

plt.tight_layout()
plt.show()
```

График представляет собой матрицу гистограмм, где каждая гистограмма соответствует распределению значений одного из признаков в наборе данных о раке груди. Это позволяет быстро оценить форму и характер распределения каждого признака.

```python
# Построение модели k-ближайших средних
k = 3  # количество соседей
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)
```

```python
# Протестировать на тестовом наборе
y_pred = model.predict(X_test)

# Вывести прогноз модели и оценить качество
print("Прогноз модели:", y_pred)
```

```python
# Оценка качества модели
accuracy = metrics.accuracy_score(y_test, y_pred)
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
```

```python
# Вывод результатов
print("Breast Cancer Dataset:")
print(f"Точность: {accuracy:.2%}")
print("Матрица ошибок:\n", confusion_matrix)
```

```python
# Визуализация матрицы ошибок
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=breast_cancer.target_names, yticklabels=breast_cancer.target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

Выводы:Модель хорошо справляется с классификацией и имеет высокую точность. Однако, стоит обратить внимание на ложные положительные и ложные отрицательные результаты. В зависимости от контекста задачи (например, если ложные отрицательные более критичны), могут потребоваться дополнительные улучшения, такие как подбор оптимальных порогов для классификации или изменение параметров модели.
