```python
# a. KNeighborsRegressor
knn_param_grid = {'n_neighbors': [3, 5, 7, 9]}
knn_grid = GridSearchCV(KNeighborsRegressor(), knn_param_grid, scoring='neg_mean_squared_error', cv=5)
knn_grid.fit(linnerud_X_train, linnerud_y_train)
best_knn = knn_grid.best_estimator_

print("Best KNeighborsRegressor Parameters:", knn_grid.best_params_)
print("Best KNeighborsRegressor MSE:", -knn_grid.best_score_)
```

```python
best_knn.fit(linnerud_X_train, linnerud_y_train)
knn_y_pred = best_knn.predict(linnerud_X_train)

plt.scatter(linnerud_y_train, knn_y_pred)
plt.title('Фактические vs Предсказанные значения для KNeighborsRegressor')
plt.xlabel('Фактические значения')
plt.ylabel('Предсказанные значения')
plt.show()
```

```python
# b. LinearRegression
linear_param_grid = {}  # Linear Regression не имеет гиперпараметров для настройки
linear_grid = GridSearchCV(LinearRegression(), linear_param_grid, scoring='neg_mean_squared_error', cv=5)
linear_grid.fit(linnerud_X_train, linnerud_y_train)
best_linear = linear_grid.best_estimator_

print("Best LinearRegression MSE:", -linear_grid.best_score_)
```

```python
# c. Ridge
ridge_param_grid = {'alpha': [0.1, 1, 10, 100]}
ridge_grid = GridSearchCV(Ridge(), ridge_param_grid, scoring='neg_mean_squared_error', cv=5)
ridge_grid.fit(linnerud_X_train, linnerud_y_train)
best_ridge = ridge_grid.best_estimator_

print("Best Ridge Parameters:", ridge_grid.best_params_)
print("Best Ridge MSE:", -ridge_grid.best_score_)
```

```python
# График для Ridge
alphas_ridge = ridge_param_grid['alpha']
mse_ridge = [-score for score in ridge_grid.cv_results_['mean_test_score']]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(alphas_ridge, mse_ridge, marker='o')
plt.xscale('log')
plt.title('Ridge: Среднеквадратичная Ошибка vs Alpha')
plt.xlabel('Alpha')
plt.ylabel('Среднеквадратичная Ошибка')

plt.tight_layout()
plt.show()
#  Графики демонстрирует изменение среднеквадратичной ошибки (MSE) в зависимости от параметра alpha.
```

```python
# d. Lasso
lasso_param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10]}
lasso_grid = GridSearchCV(Lasso(), lasso_param_grid, scoring='neg_mean_squared_error', cv=5)
lasso_grid.fit(linnerud_X_train, linnerud_y_train)
best_lasso = lasso_grid.best_estimator_

print("Best Lasso Parameters:", lasso_grid.best_params_)
print("Best Lasso MSE:", -lasso_grid.best_score_)
```
